#**4.Data Vizualization, Storytelling & Experimenting with charts: Understand the relationships between variables**

Chart -1

Question 1: Which gender spends more money and has higher purchasing power during Diwali?

```python
#Chart 1- visualization code( Gender analysis )

plt.figure(figsize=(11, 9))
sns.barplot(x='Gender', y='Amount', data=df1, estimator=sum, errorbar=None)
plt.title('Total Sales Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales (in Tens of Millions)')

plt.show()
```

*1*.Why did you pick the specific chart?

A Bar Chart with an aggregator (estimator=sum) is perfect for comparing a continuous numeric metric (Amount) across a small number of distinct categorical classes (Gender).

2.What is/are the insight(s) found from the chart?

Females contribute significantly more to the total sales amount compared to males. Females are the primary buyers in this dataset, accounting for roughly $70\%$ of the total revenue.

3.Will the gained insights help creating a positive business impact?

Are there any insights that lead to negative growth? Justify with specific reason.

Marketing budgets should lean heavily into female-centric product lines, targeted ad campaigns, and loyalty rewards tailored to female shoppers to maximize Return on Ad Spend (ROAS).
In other hand Men represent a massive underperforming segment. If the business ignores this gap, it risks stagnating in the male consumer market. Justification: The stark revenue disparity means the store is currently failing to engage male shoppers effectively, missing out on roughly half of the total potential population.

Chart - 3

Question 3 : Which geographical zones and top states generate the highest revenue?

```python
#Chart 3 - visualization (Geographic Performance : State & Zone Analysis)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Chart 1: Top 5 States by Revenue
top_states = df1.groupby('State')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(5)
sns.barplot(ax=axes[0], x='Amount', y='State', data=top_states, hue='State', palette='coolwarm', legend=False)
axes[0].set_title('Top 5 States by Total Revenue')

# Chart 2: Zone Wise Sales
zone_sales = df1.groupby('Zone')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
sns.barplot(ax=axes[1], x='Zone', y='Amount', data=zone_sales, hue='Zone', palette='viridis', legend=False)
axes[1].set_title('Total Revenue Generated by Zone')

plt.tight_layout()
plt.show()
```

*1*.Why did you pick the specific chart?

Horizontal and Vertical Bar Charts sorted in descending order make it instantly obvious to stakeholders which geographical territories are performing best.

2.What is/are the insight(s) found from the chart?

The Central and Southern Zones dominate the revenue. On a state level, Uttar Pradesh, Maharashtra, and Karnataka sit firmly at the top of the revenue leaderboards.

3.Will the gained insights help creating a positive business impact?

Are there any insights that lead to negative growth? Justify with specific reason.

Supply chain logistics, regional fulfillment centers, and localized, language-specific ad campaigns should be heavily funded in these high-ROI states to guarantee fast delivery times during the peak festive rush.

Regions like the Eastern or Northeastern zones show drastically lower sales.
Justification: Pouring blanket, non-targeted marketing funds equally across all of India will drain resources. Low-performing states will yield a negative localized ROI unless the product mix is fundamentally changed to suit regional tastes.

Question 5 : Which professional sectors possess the highest purchasing density?

Chart - 6

Question 6 : Which product categories generate the most revenue versus which sell the most volume?

```python
#Chart 6 - visualization (Product Category Performance)
fig, axes = plt.subplots(2, 1, figsize=(14, 12))

# Top Categories by Total Revenue
top_cat_rev = df1.groupby('Product_Category')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(ax=axes[0], x='Amount', y='Product_Category', data=top_cat_rev, hue='Product_Category' , palette='Blues_r', legend=False)
axes[0].set_title('Top 10 Product Categories by Total Revenue (Value)')

# Top Categories by Total Volume
top_cat_vol = df1.groupby('Product_Category')['Orders'].sum().reset_index().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(ax=axes[1], x='Orders', y='Product_Category', data=top_cat_vol, hue='Product_Category' , palette='Greens_r', legend=False)
axes[1].set_title('Top 10 Product Categories by Total Orders (Volume)')

plt.tight_layout()
plt.show()
```

*1*.Why did you pick the specific chart?

Dual Horizontal Bar Charts arranged vertically allow us to compare two different success metrics: Revenue Value vs. Order Volume.

2.What is/are the insight(s) found from the chart?

Food, Clothing & Apparel, and Electronics dominate both charts. While Food and Clothing lead heavily in volume (number of items bought), Electronics brings in massive revenue despite fewer total orders due to higher average ticket value.

Are there any insights that lead to negative growth? Justify with specific reason.

Chart - 7

Question 7 : What is the average order size, and how is individual transaction spending distributed?

```python
#Chart 7 - visualization Code (Based on: Orders & Amount)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Chart 1: Distribution of Orders per transaction
sns.countplot(ax=axes[0], x='Orders', data=df,hue='Orders' , palette='Set3', legend=False )
axes[0].set_title('Distribution of Order Quantities')

# Chart 2: Distribution of Transaction Amounts
sns.histplot(ax=axes[1], x='Amount', data=df, bins=30, kde=True, color='purple')
axes[1].set_title('Distribution of Transaction Spending')

plt.tight_layout()
plt.show()
```

*1*.Why did you pick the specific chart?

A Countplot handles the discrete, low-range nature of the Orders column, while a Histogram with a KDE (Kernel Density Estimate) curve shows the spread, skewness, and peaks of the continuous Amount column.

2.What is/are the insight(s) found from the chart?

Most transactions consist of 1 to 2 items. The spending distribution is right-skewed, showing that while most transactions hover around a moderate budget range (5,000–10,000), there is a long tail of high-value shoppers.

3.Will the gained insights help creating a positive business impact?

Are there any insights that lead to negative growth? Justify with specific reason.

Implement tier-based free shipping or discount milestones ("Add $1,500 more to your cart to unlock free shipping") to push the average transaction value higher up the distribution curve.

A massive drop-off at higher transaction values shows a lack of ultra-premium purchases.

Justification: If your platform is losing high-ticket buyers, it means your checkout flow, credit/financing options (like No-Cost EMI), or premium item inventory are not attractive enough to convince affluent shoppers to complete large transactions.
